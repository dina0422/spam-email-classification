```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("./email-spam-classification-dataset.csv", nrows=50000)
# I'm Using 50000 rows because my Machine can process it only no more
df.head()
```

```python
df.groupby('label').describe()
```

# Train/Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.text, df.label , test_size = 0.25)
```

# CountVectorizer
Builds a vocabulary of all unique tokens in the text corpus

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count[:5]
```

# MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count , y_test)
```

```python
import joblib

joblib.dump(model, "spam_classifier_model.pkl")
joblib.dump(v, "vectorizer.pkl")
```
